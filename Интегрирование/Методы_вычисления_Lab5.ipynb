
11 вариант и формула симпсона


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

# Функция по варианту
def f(x):
    return np.cos(x) / (x + 2)

def simpson(n, y, k): # Формула симпсона
    sum = y[0] + y[n-1]
    for i in range(1, n-1):
        if i % 2 == 0:
            sum += 2 * y[i]
        else:
            sum += 4 * y[i]
    return sum * k

# Функция заполняет массив x и y и получает h
def y_val(n, a, b):
    h = (b - a) / n
    x = np.linspace(a, b, n)
    y = [f(i) for i in x]
    return list(x), y, h

# Вводим диапазон X, точность eps и кол. точек
a = float(input('Введите а: '))
b = float(input('Введите b: '))
eps = float(input('Введите eps: '))
n = int(input('Введите n: '))
x, y, x2, y2 = [], [], [], []
x, y, h = y_val(n, a, b)
x2, y2, h2 = y_val(2 * n, a, b)
while abs(simpson(2 * n, y2, h2/3) - simpson(n, y, h/3)) >= eps: # Запускаем цикл, которы остановится только тогда, когда следующее знаение минус настоящее по модулю меньше eps
    n = n * 2 # Увеличиваем кол. точек в 2 раза
    x, y, h = y_val(n, a, b) # Расчитываем новые значения
    x2, y2, h2 = y_val(2*n, a, b) # Расчитываем новые значения вперед
    print(f"H = {h}")
# После выхода присваем и заново рассчитываем
n *= 2
x, y, h = y_val(n, a, b)
plt.figure(figsize=(18,6))
plt.plot(x, y)
plt.show()
s, err = integrate.quad(f, a, b) # Производим интегрирование функцией
print(f"Значение Интеграла = {s}")
print(f"Интеграл по формуле Симпсона = {simpson(n, y, h/3)}")
print(f"Погрешность = {abs(s - simpson(n, y, h/3))}")
print(f"Погрешность = {abs(s - simpson(n, y, h/3)) < eps}")
print("N = ", n)
```
