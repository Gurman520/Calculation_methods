**Функция заданная по вариантам**

```python
from sympy import diff
import numpy as np
import math
import sympy as sp
import matplotlib.pyplot as plt
```

```python
# Этот блок кода ответственен исключительно за рисование графика
def y(xi): # Задается функция по варианту
  yi = np.exp(xi + np.sin(xi))
  return yi

Y = []
X = []
print("Введите начало диапазона: ")
A = float(input())
print("Введите конец диапазона: ")
B = float(input())
print("Введите колличество точек: ")
N = int(input())
H = (B - A) / (N - 1) # Расчитывается шаг по X
print("Шаг по сетке равен: ", H)
xi = A
while xi <= B: # Один из вариантов заполнения списков значениями X и Y
  X.append(xi)
  Y.append(y(xi))
  xi += H
print("X = ", X)
print("Y = ", Y)
import matplotlib.pyplot as plt
plt.figure(1, figsize=(12, 10)) # Устанавливаем размер графика
plt.plot(X, Y) # Рисуем
plt.grid(True) # Устанавливаем сетку
print(len(X))
```

```python
# Здесь приведено решение всей задачи.
# Пример графиков будет в отдельном doc файле.
def y(xi): # Задаем свою функцию по вариантам 
  yi = np.exp(xi + np.sin(xi))
  return yi
  

# Полином Лагранжа
def lagranz(x,y,t):
  global n
  z = 0
  for i in range(n):
    p = 1
    for jet in range(0, n, 1):
      if i != jet:
        p *= (t-x[jet]) / (x[i]-x[jet])
    z += y[i] * p
  return z

# Устанавливаем начала и конец диапазона по X
A = -1.0
B = 0.5
n = 11 # Указываем количество узлов по котором будет производится интерполирование
H = 100 # Указываем количество точек по которым произвести интерполяцию

# Рисуем график график функции по 100 точкам
plt.figure(1, figsize=(12, 10))
X = np.linspace(A, B, H)
Y = [y(i) for i in X]
plt.plot(X, Y, 'r', label='My Func, 100 p.')

# Рисуем 11 узлов по которым будем вычислять
xx = np.linspace(A, B, n)
yy = [y(i) for i in xx] # Расчет функции по n точкам
plt.plot(xx, yy, "o", label='Oporn 11 p.')

# Производим интерполяцию Лагранжем по 100 точкам
xnew=np.linspace(A, B, H)
ynew=[lagranz(xx,yy,i) for i in xnew] # Расчет лагранжа 
plt.plot(xnew, ynew, 'g', label='Lagr, 100 p.')

plt.grid(True)
plt.legend() # Отвечает за добавление легенды графика
plt.show()

# Расчитываем практическую погрешность.
# Практическая погрешность это Значение Функции минус Значение Лагранжа по модолю
Rn = []
for i in range(H):
  z = abs(Y[i] - ynew[i])
  Rn.append(z) # Все погрешности заносим в массив
print(max(Rn)) # Выводим на экран макисмальную погрешность, это и будет ответ о Практической погрешности.
plt.figure(2, figsize=(10, 8)) 
plt.plot(X, Rn) # На одтельном графике рисуем график практической погрешности.



# Подробнее об этом в doc файле
def sup(xi, x, deff): # Функция нахождения максимума
  dx = sp.symbols('x') # Таким образом мы задали переменно значения символа "x"
  dif = diff(deff, dx, n + 1) # Находим n+1 производную нашей функции. n - Это количество узлов
  print(dif) 
  el = []
  for i in list(xnew): # проходим по всем значениям xnew
    el.append(dif.subs('x', i)) # находим значение выражения
  m = np.max(el) # Находим макисимально значение нашей производной
  return m

# В той же формуле есть небольшой полином. Вот его и находим.
def mult(x, rang): # rand - это массив xnew, а x - это наши узлы
  s = rang.copy()
  ss = 1
  for j in range(len(x)):
    ss = ss * (s - x[j])
  return ss

dx = sp.symbols('x')
deff = sp.exp(dx + sp.sin(dx))
su = sup(xx, xnew, deff) # Вызываем функцию нахождения супремума
fact = math.factorial(len(xx) + 1) # Расчитываем факториал n + 1
mul = np.abs(mult(xx, xnew)) # Вызываем функцию полинома
Rn = (su / fact) * mul # Находим общее значение
print("Теоретическая погрешность: ", np.max(Rn))
```
